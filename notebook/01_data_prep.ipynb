```python
import pandas as pd
import numpy as np


# create data with two columns ( name, amount)
data = {
    'product_name':np.random.choice(['Product A', 'Product B', 'Product C', 'Product D'], 100),
    'amount':np.random.normal(100, 20, 100)
}

sales_data = pd.DataFrame(data)
sales_data.head(2)
```

```python
# save the data
sales_data.to_csv('sales_data.csv', index=False)
```
